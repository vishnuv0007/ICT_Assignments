```python
#checking the columns having null values
data.isnull().sum()
```

Three columns (Age, Cabin and Embarked) are having null values

```python
#check whether Age column is normally distributed or not
data['Age'].hist()
```

```python
#Check whether Age column has outliers
sns.boxplot(data['Age'])
```

```python
#Data distribution in Age is not found to be skewed
#Also, outliers in Age column is not very far away from upper quartile value
#So, missing values are filled with mean using Simple Imputer method
from sklearn.impute import SimpleImputer
num_col=data[['Age']]
si=SimpleImputer(missing_values=np.nan,strategy='mean')
si=si.fit(num_col)
num_col=si.transform(num_col)
num_col
```

```python
Age_filled=pd.DataFrame(num_col,columns=['Age'],index=data.index)
Age_filled
```

```python
data.drop(['Age'],axis=1,inplace=True)
data
```

```python
data=pd.concat([data,Age_filled],axis=1)
data
```

```python
#Since Embarked column is categorical, its missing values are filled with mode
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
data
```

```python
data.isnull().sum()
```

Cabin column is categorical and has majority null values.
Assume that passengers belonging to one family would have occupied same cabin.
First name of passengers are found from 'Name' column and stored in another column named 'First Name'. Then this information is used to fill missing values in Cabin column.
Example: If a passenger having first name Allen is occupying a cabin B3, then all passengers with first name Allen are allotted same Cabin

```python
data['Cabin'].value_counts()
```

```python
data['First Name']=data['Name']
```

```python
name=data['First Name']
name
```

```python
#split function is used to take first name of all passengers. Note first name is seperated by a comma
pd.options.mode.chained_assignment = None
for i in range(0,891):
  name.iloc[i]=name.iloc[i].split(',')[0]
```

```python
name
```

```python
data
```

```python
#Here, a new column 'First Name' is added at end
```

```python
#Cabin column is filled using First Name
a=data['Cabin'].isnull() #seperate items in Cabin which are null values
for i in range(0,891):

  if(a.iloc[i]):   #check if that item is null
    f_name=(data.iloc[i,11:12].iloc[0])   #if yes, take First Name corresponding to that


    if(not(data[data['First Name']==f_name]['Cabin'].isnull().all())):

      s=data[data['First Name']==f_name]['Cabin'].mode()[0]  #fing mode of all cabin values corresponding to that first name

      data.iloc[i,8:9]=s  #fill cabin with above value
```

```python
data.isnull().sum()
```

```python
#It is found only 20 values got filled by this method
#Remaining values are filled with a constant string, ie, Not Available using Simple Imputer function
```

```python
from sklearn.impute import SimpleImputer
cat_col=data[['Cabin']]
si=SimpleImputer(missing_values=np.nan,strategy='constant',fill_value='Not Available')
si=si.fit(cat_col)
cat_col=si.transform(cat_col)
#cat_col
```

```python
Cabin_filled=pd.DataFrame(cat_col,columns=['Cabin'],index=data.index)
Cabin_filled
```

```python
data.drop(['Cabin'],axis=1,inplace=True)
data
```

```python
data=pd.concat([data,Cabin_filled],axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.drop('First Name',axis=1,inplace=True)
data
```

Outliers detected in 4 columns.

```python
#Outliers of 3 columns (Fare, Age and SibSp) clipped to higher quartile value
outlier_check=['Fare','Age','SibSp']
Q1=data[outlier_check].quantile(0.25)
Q3=data[outlier_check].quantile(0.75)
IQR=Q3-Q1
outliers_lower=data[outlier_check]<(Q1-1.5*IQR)
outliers_upper=data[outlier_check]>(Q3+1.5*IQR)

outliers_mild=outliers_upper | outliers_lower
data[outlier_check]=data[outlier_check].clip(upper=Q3+1.5*IQR,lower=Q1-1.5*IQR,axis=1)
data
```
