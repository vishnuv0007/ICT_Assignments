Question 1(c) - (i)  #Average Performance based on Gender

```python
#This pgm finds students of which gender (male or female) performs best on each subject based on average score.
#Avg marks obtained by males and females is calculated separately for each subject, viz, math, reading and writing.
#Then, avg of male and female of each subject are compared and results are printed.
#'\033[1;4m makes the text bold and underlined
print('\033[1;4m(i)Average Performance based on Gender\n\033[0m')
#calculate avg of math, reading and writing based on gender and results are stored in table 'avgScore' 
avgScore=student.groupby('gender').agg({'math score':['mean'],'reading score':['mean'],'writing score':['mean']}) 

#this for loop iterates through the table 'avgScore' column wise. Avg of math score of male and female is taken in 1st iteration,
#Avg of reading score of male and female is taken in 2nd iteration, Avg of writing score of male and female is taken in 3rd iteration.
#In each iteration, avg marks (in one subject) of male and female are compared and results displayed.
for i in avgScore.keys(): 
    score_female,score_male=avgScore[i].iloc[0],avgScore[i].iloc[1] #store avg marks from table 'avgScore' into two variables
    
    if(score_female>score_male):
        print("Female students got the most on average for", i[0])
    else:
        print("Male students got the most on average for", i[0])

avgScore  #display the calculated avg of math, reading and writing based on gender
```

Question 1(c) - (ii) #Average Performance based on Test Preparation Course

```python
#this pgm finds whether the students who competed test preparation course performs best on each subject based on average score.
print('\033[1;4m\n(ii)Average Performance based on Test Preparation Course\n\033[0m')
avgScore=student.groupby('test preparation course').agg({'math score':['mean'],'reading score':['mean'],'writing score':['mean']})

for i in avgScore.keys():
    score_completed,score_none=avgScore[i].iloc[0],avgScore[i].iloc[1]

    if(score_completed>score_none):
        print("Students who completed test preparation course got the most on average for", i[0])
    else:
        print("Students who do not completed test preparation course got the most on average for", i[0])

avgScore
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest

sales=pd.read_csv('Sales_add.csv')
pop_mean=sales['Sales_before_digital_add(in $)'].mean() #before sales avg
sample_mean=sales['Sales_After_digital_add(in $)'].mean()#after sales avg
sample_std_dev=sales['Sales_After_digital_add(in $)'].std() #since t test is used, sample std deviation taken in calculation
sample_size=sales['Month'].size

#Null hyp H0: There is no improvemnet in sales
#Alt hyp Ha: There is improvemnet in sales after digital marketing
#t-test chosen because sample size<30
alpha=0.05 #confidence interval taken as 95%

data=np.array(sales['Sales_After_digital_add(in $)'])
t_stats,p_value=stats.ttest_1samp(data,pop_mean,alternative="greater") #since we want to prove there is an increase in sales, one sided t test is done.
tcritical=1.721 # from one sided t table
print(t_stats,p_value)

if t_stats > tcritical:
  print("Reject null hyp: There is improvemnet in sales after digital marketing")
else:
  print("fail to reject null hyp: There is no improvemnet in sales")
```
